```python
df= pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
limit=100
df=df.iloc[0:limit,:]
df.shape
```

```python
latitude=37.77
longitude=-122.42
sfmap=folium.Map(location=[latitude, longitude], zoom_start=12)
sfmap
```

```python
incidents=folium.map.FeatureGroup()
for lat, long, in zip(df.Y, df.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat,long],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
sfmap.add_child(incidents)
```

```python
latitudes=list(df.Y)
longitudes=list(df.X)
labels=list(df.Category)
for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat,lng],popup=label).add_to(sfmap)
sfmap.add_child(incidents)
```
